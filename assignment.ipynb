## 1. Data Warehouse

Using ETL (Extract, Transform, Load) process to load data into a mongo db database.

````python
import pandas as pd
from pymongo import MongoClient

file_path = 'netflix_dataset.csv'
netflix_data = pd.read_csv(file_path)

netflix_data.head()

client = MongoClient('mongodb://test:1234@localhost:27017/')
db = client['netflix_data']

netflix_data = netflix_data.rename(columns={'Show_Id': 'id', 'Title': 'name', 'Release_Date': 'release_date'})

data_warehouse = netflix_data.to_dict(orient='records')

db['data_warehouse'].insert_many(data_warehouse)
````

## 2: Creating a Data Lake

With the goal of storing data in three collections structured, unstructured, and semi-structured data.

````python

structured_data = netflix_data.to_dict(orient='records')
db['structured_data'].insert_many(structured_data)

unstructured_data = [
    {
        'review': 'Loved the new season of Stranger Things! The storyline and characters were amazing.',
        'user': 'user123',
        'show_id': 's1'
    },
    {
        'review': 'The latest movie on Netflix was quite boring. Expected better.',
        'user': 'user456',
        'show_id': 'm1'
    },
    {
        'review': 'Just watched an amazing documentary on Netflix. Highly recommend it.',
        'user': 'user789',
        'show_id': 'd1'
    }
]
db['unstructured_data'].insert_many(unstructured_data)

semi_structured_data = [
    {
        'event': 'view',
        'timestamp': '2024-05-18T12:34:56Z',
        'details': {
            'user_id': 'user123',
            'show_id': 's1',
            'duration': '45 minutes'
        }
    },
    {
        'event': 'comment',
        'timestamp': '2024-05-18T13:14:22Z',
        'details': {
            'user_id': 'user456',
            'show_id': 'm1',
            'comment': 'Great movie! Loved the plot twists.'
        }
    }
]
db['semi_structured_data'].insert_many(semi_structured_data)
````
